```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
titanic = sns.load_dataset("titanic")
```

```python
titanic.head()
```

```python
sns.regplot( x = 'age',y='fare', data = titanic)
```

```python
sns.histplot(x = 'age', y = 'sex', hue = 'sex', stat = 'count', data = titanic);
```

```python
sns.catplot(x='pclass', y='age', hue = 'alive', kind = 'box',palette = "muted", data = titanic);
plt.grid()
```
